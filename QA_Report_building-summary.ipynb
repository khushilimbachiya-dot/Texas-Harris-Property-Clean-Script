## Replacing the '--' with null form column 'remodeled'

```python
df['remodeled'] = df['remodeled'].replace('--',np.nan)
df['remodeled']
```

```python
df['remodeled'].value_counts()
```

## Change type into intger from column 

```python
cold = ['impr_sq_ft','year_built','year','remodeled']

for col in cold:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace(',', '', regex=False)
        .str.extract(r'(\d+)', expand=False)
    )

df[cold] = df[cold].astype('Int64')
df[cold]
```
