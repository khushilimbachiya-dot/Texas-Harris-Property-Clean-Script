```python
import pandas as pd
import numpy as np

df3_small = df2[['parcelnumb']].copy()
df3_small['parcelnumb'] = df3_small['parcelnumb'].astype(str)

df_small = df[['Parcel']].copy()
df_small['Parcel'] = df_small['Parcel'].astype(str)

result = (
    df3_small.merge(
        df_small,
        left_on="parcelnumb",
        right_on="Parcel",
        how="left",
        indicator=True
    )
)

result["status"] = np.where(
    result["_merge"] == "both",
    "Matched",
    "Not Matched"
)

result = result[["parcelnumb", "status"]]

print(result)
```
