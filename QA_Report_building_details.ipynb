# Handling multi spaces

```python
from pyspark.sql import functions as F

for c, t in df.dtypes:
    if t == "string":
        df = df.withColumn(
            c,
            F.trim(
                F.regexp_replace(
                    F.col(f"`{c}`"),
                    r"\s+",
                    " "
                )
            )
        )
```

```python
from pyspark.sql import functions as F

fake_nulls = ['', 'null', 'none', 'nan', 'na', 'n/a', 'n.a.', 'nil', 'NONE']
fake_nulls_lower = [x.lower() for x in fake_nulls]

for c, dtype in df.dtypes:
    if dtype == "string":
        df = df.withColumn(
            c,
            F.when(
                F.lower(F.trim(F.col(f"`{c}`"))).isin(fake_nulls_lower),
                F.lit(None)
            ).otherwise(F.col(f"`{c}`"))
        )
```

```python
from collections import Counter
from pyspark.sql import functions as F

# ============================================================
# 1. Check for duplicate column names
# ============================================================
duplicates = [c for c, n in Counter(df.columns).items() if n > 1]

print("Duplicate columns:")
print(duplicates)

print("Total columns:", len(df.columns))
print("Unique columns:", len(set(df.columns)))

# ============================================================
# 2. List of columns to clean
# ============================================================
colv = [
    'frame_overhang_upr',
    'mas_brk_garage_pri',
    'apt_3_bedroom_unit',
    'frame_garage_pri',
    'apt_2_bedroom_unit',
    'oh_door_wood_mtl',
    'attic_unfinished',
    'office_warehouse_ratio',
    'one_story_mas_pri',
    'wall_height',
    'open_frame_porch_pri',
    'interior_finish_percent',
    'porch_open_c',
    'apt_1_bedroom_unit',
    'canopy_lwr',
    'one_story_frame_pri',
    'base_area_pri',
    'garage_interior_parking_apartments_c'
]

# ============================================================
# 3. Check which columns are missing
# ============================================================
missing = [c for c in colv if c not in df.columns]

print("\nMissing columns:")
print(missing)

# Keep only existing columns
valid_cols = [c for c in colv if c in df.columns]

print("\nColumns to be cleaned:")
print(valid_cols)

# ============================================================
# 4. Show data before cleaning
# ============================================================
print("\nData BEFORE cleaning:")
df.select(*valid_cols).show(5, truncate=False)

# ============================================================
# 5. Remove only pipe-delimited "0" values
# ============================================================
for c in valid_cols:

    cleaned = F.expr(f"""
        concat_ws(
            ' || ',
            filter(
                split(cast(`{c}` as string), '\\\\s*\\\\|\\\\|\\\\s*'),
                x -> x IS NOT NULL
                     AND trim(x) <> ''
                     AND trim(x) <> '0'
            )
        )
    """)

    df = df.withColumn(
        c,
        F.when(
            F.col(c).isNull(), None
        ).otherwise(
            F.when(cleaned == "", None)
             .otherwise(cleaned)
        )
    )

# ============================================================
# 6. Show data after cleaning
# ============================================================
print("\nData AFTER cleaning:")
df.select(*valid_cols).show(5, truncate=False)
```
